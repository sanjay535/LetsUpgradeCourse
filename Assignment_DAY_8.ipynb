```python
def armstrongNumber(num):
    temp=num
    sumOfDigitCube=0
    while num > 0:
        t=num%10
        num=int(num/10)
        sumOfDigitCube=sumOfDigitCube+t*t*t
    print(sumOfDigitCube)    
    if sumOfDigitCube == temp:
        return True
    return False
```

```python
def checkArmstrong():
    for i in range(1,1000):
        if armstrongNumber(i):
            yield i
```

```python
print(checkArmstrong())
```
